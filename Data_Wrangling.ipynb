### Handling Missing values

```python
#Replacing normalized losses column with high frequented value.
most_frequent_normalizer = data["normalized-losses"].value_counts().idxmax()
data["normalized-losses"].replace(np.nan, most_frequent_normalizer, inplace=True)
```

```python
#Replacing no of doors column with high frequented value.
most_frequent_door = data["num-of-doors"].value_counts().idxmax()
data["num-of-doors"].replace(np.nan, most_frequent_door, inplace=True)
```

```python
#replacing null values in other columns with the mean
average_bore = data["bore"].astype("float").mean(axis = 0)
data["bore"].replace(np.nan, average_bore, inplace=True)
average_stroke = data["stroke"].astype("float").mean()
data["stroke"].replace(np.nan, average_stroke, inplace = True)
average_horsepower = data["horsepower"].astype("float").mean()
data["horsepower"].replace(np.nan, average_horsepower, inplace= True)
average_peak_rpm = data["peak-rpm"].astype("float").mean()
data["peak-rpm"].replace(np.nan, average_peak_rpm, inplace= True)
average_price= data["price"].astype("float").mean()
data["price"].replace(np.nan, average_price, inplace= True)
```

The only standaridization that can be applied here are converting mpg(miles per gallon ) to kpl (km per liter)

```python
def mpg_to_kmpl(mpg):
    # Conversion factor 1 mpg (US) = 0.425144 km/L
    conversion_factor= 0.425144

    # Convert MPG to km/L
    kmpl = round((mpg * conversion_factor),2)
    return kmpl
```

```python
data["city-kmpl"] = mpg_to_kmpl(data["city-mpg"])
data["highway-kmpl"] = mpg_to_kmpl(data["highway-mpg"])
```

## 3.4 Data Normalization
> I used z- score normalization such that the length, width and height to have a mean of 0 and a standard deviation of 1.

```python

# Select the columns to transform
columns_to_normalize = ['length', 'width', 'height']


# Z-score normalization for the specified columns
mean_val = data[columns_to_normalize].mean()
std_val = data[columns_to_normalize].std()

normalized_data = (data[columns_to_normalize] - mean_val) / std_val

# Combine normalized columns with the original DataFrame
data[columns_to_normalize] = normalized_data


```


## 3.6 **Add new variable for calculating engine performance.**

>

1.   Calculate the engine capacity by multiplying the engine bore, stroke, and the number of cylinders.
2.   Creating dummy variable engine_location either front or rear.



```python
cylinder_mapping = {
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
}
```

```python
data['engine_capacity'] = (math.pi / 4) * data['bore'] ** 2 * data['stroke'] * (data['num-of-cylinders'].map(cylinder_mapping))
```

```python
indicator_var = pd.get_dummies(data["engine-location"])
indicator_var.head(5)
```

```python
# renaming the columns
indicator_var.rename(columns={"front":"engine-location-front", "rear":"engine-location-rear"}, inplace=True)
indicator_var.head(5)
```

```python
# # merge data frame and "dummy_var_1"
data = pd.concat([data, indicator_var], axis= 1)

#drop initial fuel type from data
data.drop(labels="engine-location", inplace=True, axis=1)
```
